# The Associated Press
# U.S. Shooting Incidents from 2017 to 2022

```python
incidents.head()
```

```python
incidents = incidents[incidents['date'].str.contains('|'.join(['2017', '2018', '2019', 
                                                               '2020', '2021', '2022']))]

shooting_incidents = incidents[incidents['firstcod'] == 'Shooting']
shooting_incidents.head()
```

# Offenders

```python
offenders.head()
```

```python
shooting_offenders = offenders[offenders['incident_id'].isin(shooting_incidents['incident_id'])]
shooting_offenders.head()
```

# Victims

```python
victims.head()
```

```python
shooting_victims = victims[victims['incident_id'].isin(shooting_incidents['incident_id'])]
shooting_victims.head()
```

# Weapons

```python
weapons.head()
```

```python
shooting_weapons = weapons[weapons['incident_id'].isin(shooting_incidents['incident_id'])]

shooting_weapons = shooting_weapons[shooting_weapons['weapon_type'] == 'gun']
shooting_weapons.head()
```

# Writing to CSV

```python
shooting_incidents.to_csv('../data/shooting_incidents.csv')
shooting_offenders.to_csv('../data/shooting_offenders.csv')
shooting_victims.to_csv('../data/shooting_victims.csv')
shooting_weapons.to_csv('../data/shooting_weapons.csv')
```
